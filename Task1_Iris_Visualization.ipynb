# Task 1: Exploring and Visualizing the Iris Dataset

## Objective

The goal of this task is to load, inspect, analyze, and visualize the Iris dataset using Python libraries such as Pandas, Matplotlib, and Seaborn.

## Dataset

Iris Dataset from Seaborn.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
iris = sns.load_dataset("iris")
```

## Dataset Inspection

```python
print("Shape:", iris.shape)
print("\nColumns:")
print(iris.columns)

print("\nFirst 5 Rows:")
display(iris.head())
```

```python
iris.info()
```

```python
iris.describe()
```

## Scatter Plot
Relationship between sepal length and petal length.

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=iris,
    x="sepal_length",
    y="petal_length",
    hue="species"
)

plt.title("Sepal Length vs Petal Length")
plt.show()
```

## Histograms
Distribution of numerical features.

```python
iris.hist(figsize=(12,8))
plt.tight_layout()
plt.show()
```

## Box Plot
Used to identify outliers.

```python
plt.figure(figsize=(10,6))

sns.boxplot(data=iris)

plt.title("Box Plot of Iris Features")
plt.show()
```

## Findings

1. Setosa flowers are clearly separated from other species.
2. Petal dimensions are highly useful for classification.
3. Very few outliers exist in the dataset.
4. Data appears well balanced.
